### import file

```python
import pandas as pd

sars_data = pd.read_csv('sars_2003_complete_dataset_clean.csv')

sars_data.head()
```

### what was the most deadly country?

```python
drop_date = sars_data.drop(columns='Date')

sars_by_country = drop_date.groupby(['Country']).sum()
sars_by_country['Total number of deaths'] = sars_by_country['Number of deaths']

sars_by_country.head()
```

### display in plot

```python
import matplotlib.pyplot as plt

country = [country for country, df in drop_date.groupby('Country')]
plt.figure(figsize=(15,7.5))
plt.bar(country, sars_by_country['Total number of deaths'])
_ = plt.xlabel('Country')
_ = plt.ylabel('Total number of deaths')
plt.xticks(country, rotation='vertical', size=12)
plt.show()

print('The most deadly country was China')
```

### what was the worst month in terms of number of deaths?

```python
import datetime as dt

sars_data['Date'] = pd.to_datetime(sars_data['Date']) 

sars_data['Month'] = sars_data['Date'].dt.month

sars_by_month = sars_data.groupby('Month').sum()

sars_by_month.head(50)
```

### display in plot

```python
months = (['Mar', 'Apr', 'May', 'Jun', 'Jul'])

plt.bar(months, sars_by_month['Number of deaths'], width=0.7)
plt.xticks(months)
_ = plt.xlabel('Months')
_ = plt.ylabel('Number of deaths')
plt.show()

print('The worst month was June 2003')
```

### What was the best month in terms of number recovered?

```python
months = (['Mar', 'Apr', 'May', 'Jun', 'Jul'])

plt.bar(months, sars_by_month['Number recovered'], width=0.7)
plt.xticks(months)
_ = plt.xlabel('Months')
_ = plt.ylabel('Number recovered')
plt.show()

print('The best month was June 2003')
```

### what is the join variation of the number of deaths vs. number recovered?

```python
sars_data.head()
```

```python
plt.hist2d(sars_data['Number of deaths'], sars_data['Number recovered'], bins=(10, 10), range=((0, 8), (0, 8)), cmap='coolwarm')
plt.colorbar()
plt.xlabel('Number of deaths')
plt.ylabel('Number recovered')
plt.title('Joint variation between death and recovery')
plt.show()
print('Significantly more people recovered compared to how many died')
```
