### import file

```python
import pandas as pd

sars_data = pd.read_csv('sars_2003_complete_dataset_clean.csv')

sars_data.head(50)
```

### what was the most deadly country?

```python
drop_date = sars_data.drop(columns='Date')

sars_by_country = drop_date.groupby(['Country']).sum()
sars_by_country['Total number of deaths'] = sars_by_country['Number of deaths']

sars_by_country.head()
```

### display in plot

```python
import matplotlib.pyplot as plt

country = [country for country, df in drop_date.groupby('Country')]

plt.bar(country, sars_by_country['Total number of deaths'])
_ = plt.xlabel('Country')
_ = plt.ylabel('Total number of deaths')
plt.xticks(country, rotation='vertical', size=10)
plt.show()

print('The most deadly country was, doubtless, China!!!')
```
